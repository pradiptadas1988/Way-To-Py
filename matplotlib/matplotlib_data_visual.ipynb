```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
arr_1 = np.linspace(0,6,11)
```

```python
arr_2 = arr_1 ** 2
```

```python
arr_1
```

```python
arr_2
```

```python
plt.plot(arr_1,arr_2,'r-')
```

```python
plt.plot(arr_1,arr_2)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
plt.subplot(1,3,1)
plt.plot(arr_1,arr_2)

plt.subplot(1,3,3)
plt.plot(arr_2,arr_1,'r')
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(arr_1,arr_2)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2,0.3,0.5,0.5])

axes1.plot(arr_1,arr_2,'r')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

axes2.plot(arr_2,arr_1,'b')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
my_axes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)
count = 1
for ax in axes:
    if count % 2 == 0:
        ax.plot(arr_1, arr_2, 'g')
    else:
        ax.plot(arr_2, arr_1, 'g')

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    count = count + 1

fig    
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(2,1))

axes = fig.add_axes([0,0,1,1])

axes.plot(arr_1, arr_2)
```

```python
fig = plt.figure(figsize=(6,1))

axes = fig.add_axes([0,0,1,1])

axes.plot(arr_1, arr_2)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(10,2))

axes[0].plot(arr_1, arr_2, 'r')
axes[1].plot(arr_2, arr_1, 'g')

plt.tight_layout()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(arr_1, arr_1**2, 'r',label='X Squared')
axes.plot(arr_1, arr_1**3, 'b',label='X Cubed')

axes.set_title('Legend Sample')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')

axes.legend()

fig.savefig('Plotted_Chart.png',dpi=1000)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(arr_1, arr_1**2,label='X Squared', color="purple", lw=0.5, ls='-', marker='+')
axes.plot(arr_1, arr_1**3, 'b',label='X Cubed',lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")

axes.set_title('Legend Sample')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')

axes.legend()
```
